```python
factors = [
    "Removing coordination challenges with hybrid work",
    "Increasing inclusivity",
    "Reducing headcount",
    "Increasing employee productivity",
    "Helping with repetitive/mundane tasks",
    "Increasing employee well-being",
    "Eliminating time spent on low-value activities",
    "Augmenting/enhancing human capabilities",
    "Accelerating employees' pace of work",
    "Enabling access to institutional knowledge",
    "Understanding if work aligns with company goals"
]
percentages = [16, 18, 20, 31, 29, 26, 25, 24, 24, 21, 21]

data = pd.DataFrame({'Factors': factors, 'Percentages': percentages})
data = data.sort_values(by='Percentages', ascending=True)
plt.figure(figsize=(8, 6)) 

for i, v in enumerate(data['Percentages']):
    plt.text(v + 0.3, i, str(v)+"%", color='black')


plt.barh(range(len(factors)), data['Percentages'],color='#F5A623',height=0.8)

plt.title('Factors Business Leaders Value most about AI', color='#0EA07C', fontweight='bold', fontsize=16).set_position([0.1,1.05])
plt.xlabel('Percentage')
plt.yticks(range(len(factors)), data['Factors'])
plt.tick_params(axis='y', labelsize=12, labelcolor='black')


plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)

plt.show()
```
